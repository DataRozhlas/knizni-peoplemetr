```python
def zaslovem(slovo, retezec):
    retezec = re.sub(r'[^\w\s]', '', retezec)
    for index, r in enumerate(retezec.split(' ')):
        if r == slovo:
            try:
                return retezec.split(' ')[index + 1]
            except:
                return 0
```

```python
kucharky['bez'] = kucharky['245_a'].apply(lambda x: zaslovem('bez',x))
```

```python
kucharky.explode('bez').groupby('bez').size().nlargest(20)
```

```python
kucharky['pro'] = kucharky['245_a'].apply(lambda x: zaslovem('pro',x))
kucharky.explode('pro').groupby('pro').size().nlargest(20)
```

```python
kucharky['s'] = kucharky['245_a'].apply(lambda x: zaslovem('s',x))
kucharky.explode('s').groupby('s').size().nlargest(20)
```

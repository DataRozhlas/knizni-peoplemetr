```python
x = df[['rok','245_a','072_x','650_a','655_a','700_a','700_4','700_d']]
```

```python
x = x.dropna(subset=['700_a','700_4','700_d']).explode(['700_a','700_4','700_d']).explode('245_a')
x = x.drop_duplicates(subset=['245_a','700_a'])
x = x[x['700_4'] == 'aut']
```

```python
x
```

```python
def dozili(d700):
    try:
        if "-" in d700:
            narozeni = int(re.search(r'\d{4}', d700.split('-')[0]).group(0))
            umrti = int(re.search(r'\d{4}', d700.split('-')[1]).group(0))
            vek = umrti-narozeni
        return vek
    except Exception as E:
        print(E)
        print(d700)
```

```python
x['doziti'] = x['700_d'].apply(lambda x: dozili(x))
```

```python
def meli(r, d700):
    try:
        narozeni = int(re.search(r'\d{4}', d700.split('-')[0]).group(0))
        vek = r-narozeni
        return vek
    except Exception as E:
        print(E)
        print(d700)
```

```python
x['stari'] = x.apply(lambda row: meli(row['rok'],row['700_d']), axis=1)
```

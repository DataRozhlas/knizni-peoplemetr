```python
import os
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
odkud = "downloads/sckn/zebricky"
```

```python
vsechny_knihy = []
for f in os.listdir(odkud):
    print(f)
    with open(os.path.join(odkud, f), "r", encoding="utf-8") as file:
        soup = BeautifulSoup(file.read(), "html.parser")
        zebricky = soup.find_all(class_="ranking")
        rok = re.search(r'\d{4}', soup.find('title').text).group(0)
        tyden = re.search(r'\d{1,2}', soup.find('title').text.split('č.')[1]).group(0)
        for zebricek in zebricky:
            knihy = []
            for radek in zebricek.find_all("tr"):
                if len(radek.find_all('td')) == 4:
                    kniha = {}
                    # Find all h2 tags that come before this row
                    previous_h2s = radek.find_all_previous('h2')
                    # Get the text of the first one (which is the last one before this row)
                    kategorie = previous_h2s[0].text if previous_h2s else ''
                    
                    kniha['sckn_kategorie'] = kategorie
                    kniha['rok'] = rok
                    kniha['tyden'] = tyden
                    kniha['sckn_isbn'] = radek.find_all('td')[0].text.replace('-','').strip()
                    kniha['sckn_autorstvo'] = radek.find_all('td')[1].text.strip()
                    kniha['sckn_titul'] = radek.find_all('td')[2].text.strip()
                    kniha['sckn_nakladatelstvi'] = radek.find_all('td')[3].text.strip()
                    knihy.append(kniha)
            vsechny_knihy += knihy
```

```python
pd.DataFrame(vsechny_knihy)
```

```python
pd.DataFrame(vsechny_knihy).to_json(os.path.join("data","sckn_zebricky.json"))
```

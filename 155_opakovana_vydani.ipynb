```python
df['rok'] = df['008'].apply(lambda x: najdi_rok(x))
df['pocet_stran'] = df['300_a'].apply(lambda x: pocet_stran(x))
```

```python
df.sample(20)
```

```python
df['posledni'] = df['245_a'].apply(lambda x: x[-2:])
```

```python
bordylek = df.groupby("posledni").size().nlargest(100).index.to_list()
bordylek = [x for x in bordylek if " " in x]
bordylek.append('--')
bordylek
```

```python
def bez_bordelu(pole):
    try:
        if pole != None:
            if isinstance(pole, list):
                print("seznam")
                pole = pole[0]
            pole = str(pole)
            for b in bordylek:
                if pole[-2:] == b:
                    pole = pole[:-2]
            if pole[-1] == ".":
                pole = pole[:-1]
            if pole[-1] == ",":
                pole = pole[:-1]
            return str(pole).strip()
        else:
            return ""
    except:
        print(pole)
        return ""
```

```python
df['100_a'] = df['100_a'].apply(lambda x: bez_bordelu(x))
```

```python
df['245_a'] = df['245_a'].apply(lambda x: bez_bordelu(x))
```

```python
df['podtitul'] = df['245_p'].apply(lambda x: bez_bordelu(x))
```

```python
df['titul'] = df['245_a'] + " " + df['podtitul']
df['titul'] = df['titul'].apply(lambda x: x.replace("['","/ ").replace("/']","").strip())
```

```python
import polars as pl
pl.Config(tbl_rows=200)
pl.Config.set_fmt_str_lengths(100)
```

```python
df = pl.read_parquet("data/cnb_vyber.parquet")
```

```python
df.filter(pl.col("245_a").str.contains("Věk kybernetiky"))
```

```python
def vypis(radek):
    for sloupec in radek.columns:
        if str(radek.select(pl.col(sloupec)).item()) != "None":
            print(sloupec.upper())
            informace = str(radek.select(pl.col(sloupec)).item())
            try:
                if informace.splitlines()[2] == '[':
                    informace = '\n'.join([x[2:-1].strip() for x in informace.splitlines()[3:-1]])
            except:
                pass
            print(informace)
            print("")
```

```python
vypis(df.filter(pl.col("245_a").str.contains("Věk kybernetiky")))
```

```python
df.filter(pl.col("245_c").str.contains("Encyklopedie des sciences modernes"))
```

```python
vypis(df.filter(pl.col("245_a").str.contains("Přicházíme v míru ve jménu")))
```

```python
import os
import json
import pandas as pd
```

```python
statistiky = []
pocitadlo = 0
files = [x for x in os.listdir("data/cnb_sloupce") if ".parquet" in x]
for file in files:
    pocitadlo += 1
    print(f"{pocitadlo}/{len(files)}: {file}")
    df = pd.read_parquet(os.path.join('data/cnb_sloupce', file))
    for column in df.columns.to_list():
        dfx = df.copy()
        slovnik = {}
        slovnik['sloupec'] = column
        slovnik['vyplnenost'] = len(dfx[dfx[column].notnull()]) / len(dfx)
        slovnik['delka_pred'] = len(dfx)
        dfx = dfx.explode(column)
        slovnik['delka_po'] = len(dfx)
        statistiky.append(slovnik)
```

```python
df = pd.DataFrame(statistiky)
df['lze_explodovat'] = df.apply(lambda row: True if row['delka_pred'] == row['delka_po'] else False, axis=1)
df['pomer'] = df['delka_pred'] / df['delka_po']
df['trojcisli'] = df['sloupec'].apply(lambda x: x[0:3])
```

```python
explodovatelne = df[df['lze_explodovat'] == True]['sloupec'].to_list()
explodovatelne
```

```python
with open(os.path.join('data_raw','cnb_explodovatelne_sloupce.json'), 'w+', encoding="utf-8") as file:
    file.write(json.dumps(explodovatelne))
```

```python
for trojcisli in df['trojcisli'].drop_duplicates().to_list():
    if len(df[(df['trojcisli'] == trojcisli) & (df['lze_explodovat'] == False)]) == 0:
        print(trojcisli)
```

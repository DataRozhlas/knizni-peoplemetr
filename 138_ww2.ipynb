```python
import os
import re
import statistics
from src.najdi_rok import najdi_rok
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
df = pd.read_parquet("data/cnb_vyber.parquet")
df['rok'] = df["008"].apply(lambda x: najdi_rok(x))
df = df[df['rok'].notnull()]
df = df[df['100_7'].notnull()]
```

```python
aspontri = df[df['rok'].isin(range(1918,1939))].groupby('100_7').size()
aspontri = aspontri[aspontri >= 3].index.to_list()
```

```python
zavalky = df[df['rok'].isin(range(1940,1945))]['100_7'].drop_duplicates().to_list()
```

```python
nevydavani = df[df['rok'].isin(range(1918,1939)) & (~df['100_7'].isin(zavalky))].groupby('100_7').size()
```

```python
stovky = df[['100_7','100_a']].dropna().drop_duplicates()
stovky.sample(10)
```

```python
df_nevydavani = pd.DataFrame(nevydavani).rename(columns={0:'pred_valkou'}).reset_index(drop=False).merge(stovky, on='100_7', how='left').sort_values(by='pred_valkou',ascending=False)
```

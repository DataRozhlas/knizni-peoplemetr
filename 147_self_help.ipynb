## Jak

```python
def zaslovem(slovo, retezec):
    retezec = re.sub(r'[^\w\s]', '', retezec)
    for index, r in enumerate(retezec.split(' ')):
        if r == slovo:
            try:
                return retezec.split(' ')[index + 1]
            except:
                return 0
```

```python
def pokracovani(slovicko):
    dfp = df[['245_a','rok']].copy()
    dfp[slovicko] = dfp['245_a'].apply(lambda x: zaslovem(slovicko,x))
    return dfp.explode(slovicko).groupby(slovicko).size().nlargest(20)
```

```python
pokracovani('jak')
```

```python
import os
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
slozky = [item for item in os.listdir("downloads/martinus")]
slozky
```

```python
def scrape_martinus(stranka):
    kniha = {}
    soup = BeautifulSoup(stranka, "html.parser")
    try:
        kniha['Titul'] = soup.find('h1').text.strip()
    except:
        return {}
    if soup.find('h1').find('a'):
        for odkaz in soup.find('h1').find_all('a'):
            kniha['Titul'] = kniha['Titul'].replace(odkaz.text,'').strip()
    try:    
        kniha['Autorstvo'] = [odkaz.text.strip() for odkaz in soup.find(id='author').find_all('a') if odkaz.text.strip() != 'Číst víc']
        if len(kniha['Autorstvo']) == 1:
            kniha['Autorstvo'] = kniha['autorstvo'][0]
    except:
        kniha['Autorstvo'] = soup.find(class_='product-detail__author').text.strip()
    kniha['Anotace'] = soup.find(class_='cms-article').text.strip()
    detaily = soup.find(id='details')
    for detail in detaily.find_all('dl'):
        if "série" in detail.find('dt').text.lower():
            kniha['Série'] = detail.find('dd').text.strip()
            kniha['Díl'] = re.search("\d*", detail.find('dt').text).group(0)
        else:
            if detail.find('a'):
                kniha[detail.find('dt').text.strip()] = [odkaz.text.strip() for odkaz in detail.find_all('a')]
                if len(kniha[detail.find('dt').text.strip()]) == 1:
                    kniha[detail.find('dt').text.strip()] = kniha[detail.find('dt').text.strip()][0]
            else:
                kniha[detail.find('dt').text.strip()] = detail.find('dd').text.strip()
    kniha['Cena'] = soup.find(class_='header__book').find('h3').text.strip()
    kniha['datum'] = stranka.splitlines()[-2].split(' ')[1] + ' ' + stranka.splitlines()[-2].split(' ')[2]
    flexy = soup.find_all(class_='flex-column')
    for f in flexy:
        if f.find('h6'):
            if f.find('h6').text.strip() != 'Kniha':
                kniha[f.find('h6').text.strip()] = f.find('span').text.strip()
    return kniha
```

```python
knihy = []
for s in slozky:
    odkud_brat = f'downloads/martinus/{s}'
    for i in os.listdir(odkud_brat):
        print(i)
        with open(os.path.join(odkud_brat, i), "r", encoding="utf-8") as page:
            page = page.read()
            kniha = scrape_martinus(page)
        ## kniha['kategorie_martinus'] = s
        kniha['soubor'] = i
        knihy.append(kniha)
```

```python
df = pd.DataFrame(knihy)
```

```python
import os
import re
import sys
import json
import pandas as pd
```

```python
df = pd.read_json('data_raw/ceska_beletrie_raw.json')
df = df.reindex(sorted(df.columns), axis=1)
# cizojazycne = df[(df['041_a'].notnull()) & (df['041_a'] != 'cze')]
```

```python
korpus = pd.read_csv(os.path.join("../logopedie/data","syn2000_lemma_utf8.tsv"), sep='\t', header=None)
korpus = korpus.rename(columns={0: 'rank', 1: 'lemma', 2: 'frekvence', 3: 'celková přepočítaná frekvence', 4: 'beletrie', 5: 'oborová', 6: 'publicistika', 7: 'charakteristika'})
```

```python
slova = korpus.sort_values(by="frekvence",ascending=False)['lemma'].to_list()
slova
```

```python
tituly = df[['245_a','520_a']].explode('245_a').explode('520_a')
```

```python
tituly['jednotliva_slova'] = tituly['245_a'].apply(lambda x: x[:-1].lower().replace(',','').replace('.','').replace('?','').replace('!','').strip().split(' '))
tituly['jednotliva_slova_anotace'] = tituly['520_a'].apply(lambda x: x[:-1].lower().replace(',','').replace('.','').replace('?','').replace('!','').strip().split(' ') if isinstance(x, str) else None)
```

```python
tituly = tituly.explode('jednotliva_slova')
anotace = tituly.explode('jednotliva_slova_anotace')
```

```python
uplne_vsechna = [x for x in tituly['jednotliva_slova'].to_list()]
uplne_vsechna_anotace = [x for x in anotace['jednotliva_slova_anotace'].to_list()]
```

```python
pocitadlo = 0
for s in slova:
    if pocitadlo == 50:
        break
    if s.lower() not in uplne_vsechna:
        pocitadlo += 1
        print(s)
```

```python
authors = df[df['700_4'] == 'aut'][['700_7']].copy()
authors.columns = ['author_id']

illustrators = df[df['700_4'] == 'ill'][['700_7']].copy()
illustrators.columns = ['illustrator_id']

# Merge authors and illustrators on index (book ID)
pairs = authors.join(illustrators, how='inner')

# Count frequencies of each author-illustrator pair
pair_counts = pairs.groupby(['author_id', 'illustrator_id']).size().reset_index()
pair_counts.columns = ['author_id', 'illustrator_id', 'frequency']

# Sort by frequency in descending order
pair_counts = pair_counts.sort_values('frequency', ascending=False)
```

```python
authors = df[df['700_4'] == 'aut'][['700_a']].copy()
authors.columns = ['author_id']

illustrators = df[df['700_4'] == 'ill'][['700_a']].copy()
illustrators.columns = ['illustrator_id']

# Merge authors and illustrators on index (book ID)
pairs = authors.join(illustrators, how='inner')

# Count frequencies of each author-illustrator pair
pair_counts = pairs.groupby(['author_id', 'illustrator_id']).size().reset_index()
pair_counts.columns = ['author_id', 'illustrator_id', 'frequency']

# Sort by frequency in descending order
pair_counts = pair_counts.sort_values('frequency', ascending=False)
```

```python
authors = df[df['700_4'] == 'aut'][['700_7', '700_a']].copy()
authors.columns = ['author_id', 'author_name']

illustrators = df[df['700_4'] == 'ill'][['700_7', '700_a']].copy()
illustrators.columns = ['illustrator_id', 'illustrator_name']

# Merge authors and illustrators on index (book ID)
collaborations = authors.join(illustrators, how='inner')

# Count unique authors for each illustrator
collaboration_counts = (
    collaborations.groupby(['illustrator_id', 'illustrator_name'])
    .agg({
        'author_id': 'nunique',  # Count unique authors
        'author_name': lambda x: list(set(x))  # List of unique author names
    })
    .reset_index()
)

# Rename columns for clarity
collaboration_counts.columns = [
    'illustrator_id', 
    'illustrator_name', 
    'number_of_authors',
    'author_list'
]

# Sort by number of collaborations in descending order
collaboration_counts = collaboration_counts.sort_values(
    'number_of_authors', 
    ascending=False
)
```

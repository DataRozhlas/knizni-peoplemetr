```python
import os
import json
import pandas as pd
```

```python
df = pd.read_json(os.path.join('data_raw','martinus_raw.json'))
```

```python
df.columns
```

```python
df = df.drop(columns=['Naše katalogové číslo','soubor','Filmové zpracování'])
```

```python
df[df['Kategorizace'].notnull()]['Kategorizace']
```

```python
kategorizace = df.explode('Kategorizace')
kategorizace.groupby('Kategorizace').size().nlargest(20)
```

```python
df['Ostatní'].drop_duplicates()
```

```python
df
```

```python
df['ISBN'] = df['ISBN'].apply(lambda x: str(x).split(".")[0]).astype(str)
```

```python
df['Věkové doporučení'] = df['Věkové doporučení'].apply(lambda x: x.replace("+","") if x else None)
```

```python
df
```

```python
na_int = ['Počet stran','Rok vydání','Díl','Věkové doporučení']
for i in na_int:
    df[i] = pd.to_numeric(df[i], errors='coerce', downcast='integer')
```

```python
df
```

```python
if not os.path.exists('data'):
    os.makedirs('data')
```

```python
df[df['Předběžné datum vydání'].notnull()].to_json(os.path.join('data','martinus_vyjde.json'))
df[df['Rok vydání'].notnull()].to_json(os.path.join('data','martinus_vyslo.json'))
```

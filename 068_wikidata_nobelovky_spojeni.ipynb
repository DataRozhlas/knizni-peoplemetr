```python
vsechno_nobeliststvo = wikidata_nobelovky['024_a'].to_list()
```

```python
nevysli = [x for x in vsechno_nobeliststvo if x not in cnb_nobeliststvo]
nevysli
```

```python
seznam = '|'.join(nevysli)
```

```python
url = f"https://www.wikidata.org/w/api.php?action=wbgetentities&props=labels&ids={seznam}&languages=cs|en|sk|de|fr|es|ru&format=json"
response = requests.get(url)
if response.status_code == 200:
    with open(os.path.join('downloads/wikidata/nobelovky', f"nevydani.json"), "w+", encoding="utf-8") as file:
        json.dump(response.json(), file, ensure_ascii=False, indent=4)
    print(f"Data has been saved to nevydani.json")
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
minislovnik = response.json()
```

```python
slovnik = {}
jazyky = 'cs|en|sk|de|fr|es|ru'.split('|')
for entita in minislovnik['entities']:
    labels = minislovnik['entities'][entita].get('labels')
    value = None
    for j in jazyky:
        if value == None:
            try:
                value = labels[j]['value']
            except:
                pass
    slovnik[entita] = value
```

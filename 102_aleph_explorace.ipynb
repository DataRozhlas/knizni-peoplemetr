```python
import os
import pandas as pd
```

```python
slozka = 'data_raw/aleph'
```

```python
df = pd.DataFrame()
for f in os.listdir(slozka)[0:10]:
    df = pd.concat([df, pd.read_json(os.path.join(slozka, f), dtype='str')])
```

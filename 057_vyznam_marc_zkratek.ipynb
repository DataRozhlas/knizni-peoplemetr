```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.loc.gov/marc/relators/relaterm.html")
```

```python
soup = BeautifulSoup(r.text)
```

```python
slovnik = {}
```

```python
zkratkoradky = soup.find_all("dt")
```

```python
for z in zkratkoradky:
    try:
        if z.find(class_="authorized"):
            if z.find(class_="relator-code"):
                slovnik[z.find(class_="relator-code").text.split('[')[1].split(']')[0]] = z.find(class_="authorized").text.strip()
    except:
        print(z)
```

```python
slovnik
```

```python
import json
import os
```

```python
with open(os.path.join('data','marc_zkratky.json'), 'w+', encoding="utf-8") as f:
    f.write(json.dumps(slovnik))
```

```python
import os
import pandas as pd
```

```python
df = pd.read_json(os.path.join("data","sckn_zebricky.json"))
```

```python
df
```

```python
df['roktyden'] = df.apply(lambda row: f"{row['rok']}_{row['tyden']:02}", axis=1)
```

```python
df
```

```python
df.groupby('sckn_isbn')['roktyden'].nunique().nlargest(20)
```

```python
df.groupby(['sckn_autorstvo','sckn_titul'])['roktyden'].nunique().nlargest(20)
```

```python
df[df['rok'] >= 2020].groupby(['sckn_autorstvo','sckn_titul'])['roktyden'].nunique().nlargest(60)
```

```python
df[df['rok'] >= 2020].groupby('sckn_nakladatelstvi').size().nlargest(60)
```
